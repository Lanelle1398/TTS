### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
````

````python
# Exercise 1

import pandas as pd

company_sales = pd.read_csv("data/company_sales_data.csv")
# company_sales.describe()
company_sales
````

````python
plt.plot(company_sales[ "month_number"], company_sales["total_profit"])
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.show()
````

### Exercise 2. 

Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
#Line Style dotted, Line-color should be red, circle marker, line width as 3
plt.plot(company_sales["month_number"],company_sales["facecream"], linestyle='dashed', color= 'red', marker='o', linewidth=3)
plt.xlabel("Month number")
plt.ylabel("Sold units number")

#place legend in lower right of the plot
plt.legend(loc='lower right', title='total profit over time')
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3

# Plotting both the curves simultaneously
plt.plot(company_sales["month_number"],company_sales["facecream"], linestyle='dashed', color= 'grey', linewidth=2, label="facecream")
plt.plot(company_sales["month_number"],company_sales["facewash"], linestyle='dashed', color= 'blue', linewidth=2, label="facewash")
plt.plot(company_sales["month_number"],company_sales["toothpaste"], color= 'red', linewidth=2, label="toothpaste")
plt.plot(company_sales["month_number"],company_sales["bathingsoap"], linestyle='dashed', color= 'green', linewidth=2, label="bathingsoap")
plt.plot(company_sales["month_number"],company_sales["shampoo"], linestyle='dashed', color= 'purple', linewidth=2, marker='o', label="shampoo")
plt.plot(company_sales["month_number"],company_sales["moisturizer"], linestyle='dashed', color= 'orange', linewidth=2, marker='v', label="moisturizer")
plt.legend()
plt.title("All Product Sales")
plt.xlabel("Month number")
plt.ylabel("Sold units number")
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
# Exercise 5
fig = plt.figure()
# ax = fig.add_axes([0,0,1,1])
plt.bar(company_sales["month_number"], company_sales["facewash"], color='red', width=0.20, label="facewash")
plt.bar(company_sales["month_number"] + .20, company_sales["facecream"], color='blue', width=0.20, label="facecream")
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.legend()
plt.title("Exercise 5")
````
