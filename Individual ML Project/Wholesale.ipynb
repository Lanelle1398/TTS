### Can I use k-means clustering to find the spending patterns of clients of a whole sale distributor on diverse product categories?

Choosing optimal number of clusters for K-Means through Elbow method 

finding wcss value for different number of clusters

```python
#for fresh and channel
wcss=[]
for i in range(1,10):
        kmeans=KMeans(n_clusters=i,init='k-means++', random_state=20)
        kmeans.fit(X)
        
       wcss.append(kmeans.inertia_)
# for milk and channel
wcss1=[]
for i in range(1,10):
        kmeans=KMeans(n_clusters=i,init='k-means++', random_state=20)
        kmeans.fit(X1)
        
        wcss1.append(kmeans.inertia_)
# for grocery and channel
wcss2=[]
for i in range(1,10):
        kmeans=KMeans(n_clusters=i,init='k-means++', random_state=20)
        kmeans.fit(X2)
        
        wcss2.append(kmeans.inertia_)
wcss3=[]
for i in range(1,10):
        kmeans=KMeans(n_clusters=i,init='k-means++', random_state=20)
        kmeans.fit(X3)
        
        wcss3.append(kmeans.inertia_)
        
wcss4=[]
for i in range(1,10):
        kmeans=KMeans(n_clusters=i,init='k-means++', random_state=20)
        kmeans.fit(X4)
        
        wcss4.append(kmeans.inertia_)
wcss5=[]
for i in range(1,10):
        kmeans=KMeans(n_clusters=i,init='k-means++', random_state=20)
        kmeans.fit(X5)
        wcss5.append(kmeans.inertia_)
```

```python
#plot an elbow graph
sns.set()
plt.plot(range(1,10),wcss)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.title('Fresh')
plt.show()

#plot an elbow graph
sns.set()
plt.plot(range(1,10),wcss1)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.title('Milk')
plt.show()

#plot an elbow graph
sns.set()
plt.plot(range(1,10),wcss2)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.title('Grocery')
plt.show()

#plot an elbow graph
sns.set()
plt.plot(range(1,10),wcss3)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.title('Frozen')
plt.show()

#plot an elbow graph
sns.set()
plt.plot(range(1,10),wcss4)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.title('Detergents_paper')
plt.show()


#plot an elbow graph
sns.set()
plt.plot(range(1,10),wcss5)
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.title("Delicassen")
plt.show()
```

Optimum number of clusters for fresh is 5.

Optimum number of clusters for grocery is 6.

Optimum number of clusters for frozen is 6.

Optimum number of clusters for delicassen is 5.

### Analysis

The point of this project was to use clustering to figure out the spending patterns of hotel,retail, cafe and others.
By knowing the spending patterns of clients, companies can make changes to their stores to earn more from their most profitable clients.
We first preprocess our data by loading and checking for null values.

Then, I find the optimal number of clusters for each category(fresh, milk, grocery, frozen, detergent_paper and delicassen) and channel.
Then I predict the cluster that all the values  belong to for each category.

Then for channel 1 or 2, I plot the total values spent on the category. I put these values in clusters.

From our scatter plots, we see that hotel, retail, and cafe spends more on fresh and frozen foods than 'other' clients.By eyeballing the scatterplots, we incorrectly infer that other channels spend more on delicassen, when in actuality, hotels, retails and cafe spend more on delicassen.There might be an extreme outlier in hotels, retails and cafe that would require further investigation.


Other clients spend more on  detergent_papers, grocery and milk than  hotels, retails and cafes.By looking at the scatterplot for milk, one can self-evidently infer that 'other' spend more on milk than hotels, retails, and cafes due to the outlier.

To summarize our findings, hotels, retails and cafes(channel 1) spent more on fresh, frozen and  delicassen. Other(channel 2) spent more on detergent_paper, grocery and milk than.
